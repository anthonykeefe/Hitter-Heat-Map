```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# Create custom colormap
colors = ["#3e66cc", "#ffffff", "#c90000"]
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

# Assuming pitcher_data is a DataFrame with columns 'PlateLocSide' and 'PlateLocHeight'
data = pd.read_csv("/Users/anthonykeefe/Desktop/all_wareham_games.csv", low_memory=False)

# Ensure columns match column names 
hitter_column = 'Batter'
hit_x_column = 'PlateLocHeight'
hit_y_column = 'PlateLocSide'
exit_velocity_column = 'ExitSpeed'
batter_team_column = 'BatterTeam'

# Filter 90+ mph exit velocity
data_filtered = data[data[exit_velocity_column] > 90]

# Filter Gatemen Players only
data_team = data[data[batter_team_column] == 'WAR_GAT']

# Get a list of unique hitters from filtered data
hitters = data_team[hitter_column].unique()

# Create a heat map for each hitter
for hitter in hitters:
    # Filter the data for the current hitter
    hitter_data = data_team[data_team[hitter_column] == hitter]

    # Skip hitters with less than a certain number of data points (e.g., 5)
    if len(hitter_data) < 5:
        print(f"Skipping {hitter} due to insufficient data points.")
        continue

    # Invert PlateLocSide to reflect the hitter's perspective
    hitter_data[hit_y_column] = -hitter_data[hit_y_column]

    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(8, 8))

    # Plot the 2D density
    sns.kdeplot(
        data=hitter_data,
        x=hit_x_column,
        y=hit_y_column,
        fill=True,
        cmap=cmap,
        ax=ax,
        thresh=0
    )

    # Customize the plot
    ax.set_title(f'Heat Map of Hit Locations for {hitter} (Exit Velocity > 90 mph)')
    ax.set_xlabel('PlateLocHeight (Vertical)')
    ax.set_ylabel('PlateLocSide (Horizontal)')
    ax.set_facecolor('white')
    fig.patch.set_facecolor('white')
    ax.grid(False)

    # Optionally, save the plot as an image file
    plt.savefig(f'heatmap_{hitter}_over_90mph.png')

    # Display the plot
    plt.show()
```
